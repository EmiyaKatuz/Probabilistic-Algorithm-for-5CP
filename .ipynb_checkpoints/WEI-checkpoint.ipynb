## Motivation

Traditional error metrics have notable limitations. Standard measures like Mean Absolute Error (MAE) and Root Mean Square Error (RMSE) treat errors at all times equally, regardless of load level. This is problematic because errors during peak-demand hours carry far greater operational and economic risk than errors during low-demand periods. For instance, the top five peak hours of the year often determine capacity planning costs, under forecasting these peaks can incur substantial penalties. However, an MAE would average such peak errors out with many trivial off-peak errors, obscuring their impact. RMSE, by squaring errors, does give more weight to large errors; yet it still does not explicitly differentiate whether a large error occurred on a high-demand day or a moderate day. In practice, a forecasting model could achieve a low RMSE by overall accuracy yet still consistently miss the absolute highest peaks.

Another limitation of conventional metrics is seen with percentage errors like Mean Absolute Percentage Error (MAPE). MAPE inherently underweight's errors during high-demand hours: an absolute error of 500 MW when the actual load is 20000 MW is only 2.5% MAPE, whereas the same 500 MW miss on a 5000 MW load is 10%. Thus, MAPE can paradoxically make a peak under forecast appear less significant in percentage terms, even though its real impact is larger. WEI is designed to ensure that forecast errors during high-demand periods are appropriately emphasized in model evaluation.

## Math

WEI is defined as a weighted mean absolute error over the time horizon of interest, where the weights are a function of the demand level. Let $A_t$ be the actual load at time $t$, $P_t$ the predicted (forecasted) load at time $t$, and $w_t$ a non-negative weight assigned to the error at time $t$. The general formula for WEI over $T$ time intervals is:

$$\mathrm{WEI}=\frac{\sum_{t=1}^T w_t\lvert P_t-A_t\rvert}{\sum_{t=1}^T w_t}$$

This formula produces a weighted average of the absolute errors, in which each error $|P_t-A_t|$ is scaled by $w_t$. By choosing $w_t$ to be larger for times of higher demand, WEI will upweight errors when $A_t$ is large, effectively concentrating the accuracy metric on high-load periods. A natural choice is to set $w_t$ proportional to the actual load itself. For example, one could choose $w_t = A_t$. In that case, the numerator becomes $\sum A_t|P_t - A_t|$, and the denominator $\sum A_t$. This can directly reflect the absolute significance of the error in terms of load served.

Or we can set a baseline weight of 1 for all times, but if $A_t$ exceeds a chosen percentile (say the 95th percentile of demand), assign a larger weight $w_t=\alpha>1$ for those times. In practice, this could mean all hours in the top 5\% of load get an extra penalty factor. For example, $\alpha=3$ so that errors during the most critical peak hours count triple.

## Ontario

For this example, let’s choose two weighting schemes to illustrate:

(1) proportional weighting $w_t = A_t$

(2) threshold weighting where:

$$w_t= \begin{cases}1, & A_t<A_{0.95} \\ K, & A_t \geq A_{0.95}\end{cases}$$

with $A_{0.95}$ being the 95th percentile of historical demand, and $K$ some constant > 1 (for example, $K=4$ to quadruple the weight on top 5\% demand hours). Scheme (1) will smoothly weight every hour by its demand level, whereas scheme () will sharply distinguish peak hours from others, effectively isolating the extreme peak error contribution.

For each time $t$, compute the absolute error $E_t = |P_t - A_t|$. Then compute the weighted error $W_t = w_t \cdot E_t$ using the chosen $w_t$. This yields, for every timestamp, the contribution to the numerator of WEI.

Sum up all weighted errors and divide by the sum of weights:

$$\mathrm{WEI}=\frac{\sum_t W_t}{\sum_t w_t}=\frac{\sum_t w_t\lvert P_t-A_t\rvert}{\sum_t w_t}$$

The result of this computation is a single number with the same unit as the error. For example, suppose the day-ahead forecast over 2022–2024 has an MAE of 200 MW. After applying weight = actual, we might get a WEI = 300 MW, meaning on average (weighted by load), the error is 300 MW. This indicates that more errors are occurring during high-load periods. If WEI had instead been equal to MAE, it would mean errors are evenly distributed regardless of demand level. A much larger WEI than MAE signals that the errors are skewed to high-demand times.

```python
# ----------------- 1. Imports & Parameters -----------------
import pandas as pd
import numpy as np
from pathlib import Path
from datetime import timedelta

DATA_DIR = Path(".")
FILES = [DATA_DIR / f"IESO-Ontario Demand 2022-2025 {n} day ahead.csv" for n in range(1, 7)]

# weight parameters
THRESH_PCTL = 0.95  # high‑load threshold (95th percentile)
THRESH_FACTOR = 4  # weight multiplier above threshold
```

Assuming our target is to evaluate the time period in summer, from June to September, then it will be 4 months with 4 CP days each, so 16 days in total. The proportion of the year is 16/365 = 0.0438, or 4.38%, which is close to the 5% threshold. So it is reasonable to use the 95th percentile of demand as a threshold for peak hours.

```python
# ----------------- 2. Time parser ------------------
def safe_timestamp(row):
    """Handle '24:00:00' by rolling to next‑day 00:00:00."""
    date_str = str(row['Date']).strip()  # e.g. '1/01/2022'
    time_str = str(row['Time']).strip()  # e.g. '24:00:00'

    # Parse the date (dayfirst=True for d/m/yyyy)
    base = pd.to_datetime(date_str, dayfirst=True, errors="raise")
    # Split time into components
    parts = list(map(int, time_str.split(":")))
    if len(parts) == 2:  # HH:MM
        h, m = parts
        s = 0
    elif len(parts) == 3:  # HH:MM:SS
        h, m, s = parts
    else:
        raise ValueError(f"Unexpected Time format: {time_str}")

    # Wrap 24:00 to next‑day 00:00
    if h == 24:
        base += timedelta(days=1)
        h = 0

    return base + timedelta(hours=h, minutes=m, seconds=s)
```

```python
# ----------------- 3. Load & merge ------------------------
def load_concat(files):
    dfs = []
    for fp in files:
        if not fp.exists():
            raise FileNotFoundError(fp)
        df = pd.read_csv(fp)
        # Standardise column names
        df.columns = [c.strip() for c in df.columns]
        # rename key columns for consistency
        rename_map = {
            "ECA: IESO-Ontario Demand Historic Forecast": "ECA",
            "RTO: IESO-Ontario Demand Historic Forecast": "RTO",
            "TESLA: IESO-Ontario Demand Historic Forecast": "TESLA",
            "TESLA: IESO-Ontario Demand Actual": "Actual",
        }
        df = df.rename(columns=rename_map)
        # create a timestamp
        df['timestamp'] = df.apply(safe_timestamp, axis=1)
        dfs.append(df[["timestamp", "ECA", "RTO", "TESLA", "Actual"]])
    return pd.concat(dfs, ignore_index=True)


df = load_concat(FILES)
```

```python
# ----------------- 4. WEI function -------------------------
def compute_wei(df, model, scheme="threshold"):
    actual = df["Actual"]
    pred = df[model]
    abs_err = (pred - actual).abs()

    if scheme == "proportional":
        weights = actual
    elif scheme == "threshold":
        thr = actual.quantile(THRESH_PCTL)
        weights = np.where(actual >= thr, THRESH_FACTOR, 1.0)
    else:
        raise ValueError("Unknown scheme")

    wei = (weights * abs_err).sum() / weights.sum()

    # extra: share of weighted error from top 5% demand
    top_mask = actual >= actual.quantile(THRESH_PCTL)
    share = (weights[top_mask] * abs_err[top_mask]).sum() / (weights * abs_err).sum()

    return wei, share
```

```python
# ----------------- 5. Summary ----------------------
models = ["ECA", "RTO", "TESLA"]
rows = []
details = []
for m in models:
    wei_p, share_p = compute_wei(df, m, "proportional")
    wei_t, share_t = compute_wei(df, m, "threshold")
    rows.append([m, round(wei_p, 2), round(wei_t, 2)])
    details.append([m, round(share_p * 100, 2), round(share_t * 100, 2)])

wei_df = pd.DataFrame(rows, columns=["Model", "WEI_proportional", "WEI_threshold"])
share_df = pd.DataFrame(details, columns=["Model",
                                          "PeakShare_proportional(%)",
                                          "PeakShare_threshold(%)"])
wei_df, share_df
```

### Meaning of the results:

**WEI_proportional**: Weighted by actual load size for each time period. The lower the value, the less the model errs during ‘high load’ periods across the entire load range.

**WEI_threshold**: Amplifies the weight of time periods exceeding the 95th percentile of an annual load by 4 times. This value specifically measures prediction errors during ‘peak hours.’ The lower the value, the more reliably the model captures extreme peaks.

The difference between the two sets of numbers = the additional amplified error during peak hours. If the difference is large, it indicates that the model has significant prediction deviations during peak hours.

### Meaning of the share values:

**PeakShare_proportional**

This is computed as follows:
1. Multiply every hourly error by its actual load.
2. Restrict to the top 5 % highest-load hours.
3. Divide the weighted error in those hours by the weighted error in all hours.	Share of total weighted error that comes from the busiest 5 % hours.

For example, if this number is 40 %, almost half of all “load-weighted mistakes” are concentrated in that tiny slice of time.

**PeakShare_threshold**

Same idea, but now the weights are 1 below the 95th percentile and 4x above it.

The same share, but after aggressively amplifying the importance of peak hours. This highlights how badly a model could hurt if we explicitly care four times more about errors at system peaks.

**So,**

A low PeakShare (10–15 %) means that weighted errors are relatively balanced across the year; the model is not systematically failing at peaks.

A high PeakShare (35–40 %) signals that a disproportionate fraction of load-weighted error occurs in the very hours we care about most.

The threshold version exaggerates this contrast. If a model’s PeakShare jumps from 17 % (proportional) to 39 % (threshold), we learn that once peak hours are given extra importance, the model’s deficiencies at those times dominate its overall performance score.

### Performance Summary:

The WEI of RTO/TESLA is approximately half that of ECA, indicating that their average absolute error in high-load scenarios is significantly smaller. RTO and TESLA have identical values, indicating that their prediction sequences are nearly identical within the sample period.

### Peak Sensitivity:

- ECA: From 745 to 865 MW, an increase of approximately +16%. This indicates that once the weighting for peak hours is specifically amplified, the error significantly increases, with the most severe inaccuracy occurring during extreme peak periods.

- RTO/TESLA: Only from 379 to 392 MW, an increase of approximately +3%, indicating seamless adaptation to weight amplification, suggesting a more robust prediction of peak demand.

#### High-load error proportion:

In the load-weighted WEI, the top 5% of high-load hours in ECA account for 39% of the weighted error; RTO/TESLA only 25%. In other words, ECA's errors are highly concentrated during the system's most critical peak periods, while RTO/TESLA's errors are more evenly distributed, resulting in greater accuracy during key peak periods.


```python
# ----------------- 6. Plot WEI bar chart -------------------
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
x = np.arange(len(models))
plt.bar(x - 0.15, wei_df["WEI_proportional"], width=0.3, label="Proportional")
plt.bar(x + 0.15, wei_df["WEI_threshold"], width=0.3, label="Threshold‐4×")
plt.xticks(x, models)
plt.ylabel("WEI (MW)")
plt.title("Demand‑Weighted Error Index comparison")
plt.legend()
plt.tight_layout()
plt.show()
# --------------- 7. Plot peak‑error share ------------------
plt.figure(figsize=(6, 4))
plt.bar(models, share_df["PeakShare_threshold(%)"], color="salmon")
plt.ylabel("Weighted error share in top‑5% load hours (%)")
plt.title("How much of total weighted error comes\nfrom the highest‑load 5% hours?")
plt.tight_layout()
plt.show()
```

The plot below shows ‘how the weighted errors of the three models are concentrated in the load distribution, starting from the peak load hour’:

**X-axis**: Cumulative percentage of hours (left end = peak hour, right end = lowest load hour)

**Y-axis**: Cumulative weighted error percentage

The steeper the curve, the more the model's errors are concentrated in a small number of peak hours; a flatter curve indicates a more even distribution of errors.

### Load distribution

Imagines lining up every hour of the year from the busiest (the highest demand) on the left to the quietest (the lowest demand) on the right. This ordered line is the load distribution.

### Cumulative weighted error

For each hour we first measure the absolute forecast error (how many megawatts the model missed by).

We then weight that error by the actual demand in that hour.

  • If the system was carrying 25 000 MW, the same 100 MW miss is given 25 000 units of importance.

  • If the system was only at 10 000 MW, that same 100 MW miss is given 10 000 units.

These weighted errors are added up hour-by-hour from left to right; the running total is the cumulative weighted error curve.

If the curve rises very steeply in the first few percent of hours, it tells us that most of the model’s serious mistakes occur precisely when demand is highest.

```python
#--------------- 8. Plot cumulative weighted error curve ------------------
df_sorted = df.sort_values("Actual", ascending=False).reset_index(drop=True)
hours_pct = np.arange(1, len(df_sorted) + 1) / len(df_sorted)

curves = {}
for m in models:
    abs_err = (df_sorted[m] - df_sorted["Actual"]).abs()
    w_err = df_sorted["Actual"] * abs_err
    curves[m] = np.cumsum(w_err) / w_err.sum()

plt.figure(figsize=(6, 4))
for m in models:
    plt.plot(hours_pct * 100, curves[m] * 100, label=m)
plt.xlabel("Top‑load hours (cumulative %)")
plt.ylabel("Cumulative % of weighted error")
plt.title("Where do weighted errors concentrate?")
plt.gca().invert_xaxis()
plt.grid(alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

## Actual vs. Forecasts around the annual peak

The plot below shows the actual load and the forecasts around the annual peak. The annual peak is defined as the highest load in the year, and the plot shows a 7-day window around that peak. The x-axis represents the timestamp, and the y-axis represents the load in MW. The plot includes the actual load, the ECA forecast, the RTO forecast, and the TESLA forecast. The actual load is shown in blue, the ECA forecast in orange, the RTO forecast in green, and the TESLA forecast in red.

```python
# Select a 7‑day window around the annual peak
peak_day = df.loc[df["Actual"].idxmax(), "timestamp"].normalize()
window = (df["timestamp"] >= peak_day - pd.Timedelta(days=2)) & \
         (df["timestamp"] <= peak_day + pd.Timedelta(days=4))

subset = df.loc[window].set_index("timestamp")

subset[["Actual","ECA","RTO","TESLA"]].plot(figsize=(10,4))
plt.title("Actual vs. Forecasts around annual peak")
plt.ylabel("Load (MW)"); plt.tight_layout(); plt.show()
```
